```python
def distinct_numbers_counter(a, b):

    numbers_list = []
    final_list = []
    
    for i in range(2, a + 1):
        for j in range(2, b + 1):
            numbers_list.append(i ** j)

    sorted_numbers_list = sorted(numbers_list)

    for i in sorted_numbers_list:
        counter = numbers_list.count(i)
        if counter == 1:
            final_list.append(i)

        else:
            numbers_list.pop(numbers_list.index(i))

    print(f"The distinct numbers are: {len(final_list)}")
    return sorted(final_list)

distinct_numbers_counter(100, 100)
```
