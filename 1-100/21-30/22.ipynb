```python
with open("names.txt", "r") as file:
    list_of_names = file.readlines()
    string_of_names = list_of_names[0]

list_of_names = string_of_names.split(",")

for i in range(len(list_of_names)):
    list_of_names[i] = list_of_names[i].replace('"', '')
    

alphabet_dictionary = {
    "A" : 1,
    "a" : 1,
    "B" : 2,
    "b" : 2,
    "C" : 3,
    "c" : 3,
    "D" : 4,
    "d" : 4,
    "E" : 5,
    "e" : 5,
    "F" : 6,
    "f" : 6,
    "G" : 7,
    "g" : 7,
    "H" : 8,
    "h" : 8,
    "I" : 9,
    "i" : 9,
    "J" : 10,
    "j" : 10,
    "K" : 11,
    "k" : 11,
    "L" : 12,
    "l" : 12,
    "M" : 13,
    "m" : 13,
    "N" : 14,
    "n" : 14,
    "O" : 15,
    "o" : 15,
    "P" : 16,
    "p" : 16,
    "Q" : 17,
    "q" : 17,
    "R" : 18,
    "r" : 18,
    "S" : 19,
    "s" : 19,
    "T" : 20,
    "t" : 20,
    "U" : 21,
    "u" : 21,
    "V" : 22,
    "v" : 22,
    "W" : 23,
    "w" : 23,
    "X" : 24,
    "x" : 24,
    "Y" : 25,
    "y" : 25,
    "Z" : 26,
    "z" : 26,
}

name_number_dict = {}

for name in list_of_names:
    sum = 0
    for char in name:
        sum += alphabet_dictionary[char]
    name_number_dict[name] = sum

sorted_name_number_dict = dict(sorted(name_number_dict.items()))
index_dict = {}

for i, name in enumerate(sorted_name_number_dict):
    index_dict[name] = i + 1

sum = 0
for name in sorted_name_number_dict:
    sum += sorted_name_number_dict.get(name) * index_dict.get(name)

print(sum)

```
